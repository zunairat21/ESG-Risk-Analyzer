```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

#Selecting only numeric features for VIF Calaculation
numeric_cols=data.select_dtypes(include=['float64', 'int64']).drop(columns=['CompanyID', 'Year'])
X=add_constant(numeric_cols)

#Compute VIF for each feature
vif_data=pd.DataFrame()
vif_data['Feature']=X.columns
vif_data['VIF']=[variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

#Dispaly VIF values
vif_data.sort_values(by='VIF', ascending=False)
```

```python
#CarbonEmissions, WaterUsage, and EnergyConsumption are highly correlated (VIF > 300).
#Keeping all 3 would inflate multicollinearity and confuse the model.
data.drop(columns=['CarbonEmissions', 'WaterUsage'], inplace=True)
display(data)
```

```python
# Phase 04: Feature Selection
data.drop(columns=['CompanyID', 'CompanyName', 'Year',  'EnergyConsumption'], inplace=True)
```

```python
# Phase 05: Regression Model - Predict ESG_Overall (Risk Score)
# Split Features & Target for Regression
X_reg = data.drop(columns=['ESG_Overall'])
y_reg = data['ESG_Overall']
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
# Identify column types
num_cols = X_reg.select_dtypes(include=['float64', 'int64']).columns
cat_cols = X_reg.select_dtypes(include=['object']).columns

# Build Preprocessing Pipeline
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])
from joblib import dump
dump(preprocessor, 'preprocessor.joblib')
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, classification_report
from joblib import dump
# Full Pipeline for Regression
reg_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(random_state=42))
])
```

```python
from sklearn.model_selection import train_test_split
# Train Regression Model
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
reg_pipeline.fit(X_train_reg, y_train_reg)
```

```python
import numpy as np
# Evaluate
y_pred_reg = reg_pipeline.predict(X_test_reg)
print(f"RMSE: {np.sqrt(mean_squared_error(y_test_reg, y_pred_reg)):.2f}")
```

```python
import shap
# SHAP Explainability - Regression
reg_model = reg_pipeline.named_steps['model']
X_sample_reg = reg_pipeline.named_steps['preprocessor'].transform(X_test_reg[:100])
if hasattr(X_sample_reg, 'toarray'):
    X_sample_reg = X_sample_reg.toarray()
feature_names_reg = [name.split("__")[-1] for name in reg_pipeline.named_steps['preprocessor'].get_feature_names_out()]
explainer_reg = shap.Explainer(reg_model)
shap_values_reg = explainer_reg(X_sample_reg)
shap.summary_plot(shap_values_reg, X_sample_reg, feature_names=feature_names_reg, plot_type="bar", plot_size=(10, 6), max_display=15)
```

```python
# Save model
dump(reg_pipeline, 'rf_regression_pipeline.joblib')
```

```python
# Phase 06: Create Classification Label (ESG Risk Level)
def bin_esg_risk(score):
    if score > 80:
        return 'Low_Risk'
    elif score > 50:
        return 'Medium_Risk'
    elif score > 20:
        return 'High_Risk'
    else:
        return 'Very_High_Risk'

data['ESG_Risk_Level'] = data['ESG_Overall'].apply(bin_esg_risk)
```

```python
# Phase 07: Classification Model - Predict Risk Level
X_cls = X_reg.copy()  # Same features as before
y_cls = data['ESG_Risk_Level']
```

```python
# Encode target
label_encoder = LabelEncoder()
y_cls_encoded = label_encoder.fit_transform(y_cls)
```

```python
# Print class mappings
class_mappings = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("\n✅ Class label encoding:")
for k, v in class_mappings.items():
    print(f"{k} → {v}")

# ✅ Save the label encoder
from joblib import dump
dump(label_encoder, 'label_encoder.joblib')
print("✅ Label encoder saved.")
```

```python
# Train-Test Split
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls_encoded, test_size=0.2, random_state=42)
```

```python
# Step: Preprocess Training Data
X_train_cls_processed = preprocessor.fit_transform(X_train_cls)
```

```python
from imblearn.over_sampling import SMOTE
# Step: Apply SMOTE
smote = SMOTE(random_state=42)
X_train_cls_resampled, y_train_cls_resampled = smote.fit_resample(X_train_cls_processed, y_train_cls)
```

```python
from lightgbm import LGBMClassifier
# Train LightGBM Classifier
clf_model = LGBMClassifier(random_state=42)
clf_model.fit(X_train_cls_resampled, y_train_cls_resampled)
```

```python
# Process X_test_cls using the preprocessor
X_test_cls_processed = preprocessor.transform(X_test_cls)

# Get correct feature names from preprocessor
from sklearn.compose import ColumnTransformer
import pandas as pd

if hasattr(preprocessor, 'get_feature_names_out'):
    feature_names = preprocessor.get_feature_names_out()
else:
    # Fallback: make up generic names if needed
    feature_names = [f"feature_{i}" for i in range(X_test_cls_processed.shape[1])]

# Wrap the transformed data back into a DataFrame
X_test_cls_df = pd.DataFrame(X_test_cls_processed, columns=feature_names)

# Predict using LightGBM (✅ with named columns)
y_pred_cls = clf_model.predict(X_test_cls_df)

# Print or evaluate
from sklearn.metrics import classification_report
print(classification_report(y_test_cls, y_pred_cls, target_names=label_encoder.classes_))
```

```python
# Save trained components
dump(clf_model, 'lgb_model.joblib')
print("✅ Model saved successfully.")
```

```python
# SHAP Explainability - Classification
X_sample_cls = X_test_cls[:100]
X_sample_cls_transformed = preprocessor.transform(X_sample_cls)
if hasattr(X_sample_cls_transformed, 'toarray'):
    X_sample_cls_transformed = X_sample_cls_transformed.toarray()
explainer_cls = shap.Explainer(clf_model.predict_proba, X_sample_cls_transformed)
shap_values_cls = explainer_cls(X_sample_cls_transformed)
feature_names_cls = [name.split("__")[-1] for name in preprocessor.get_feature_names_out()]
```
